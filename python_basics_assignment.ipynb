```python
class Shape:
    def area(self):
        return 0

class Square(Shape):
    def area(self):
        return 25

s = Square()
print("Polymorphism – Shape:", s.area())
```

```python
class Car:
    def start(self):
        return "Car starts"

class Bike:
    def start(self):
        return "Bike starts"

def show_start(vehicle):
    print(vehicle.start())

print("\nPolymorphism – Car/Bike:")
show_start(Car())
show_start(Bike())
```

```python
class Animal:
    def sound(self):
        return "Some sound"

class Dog(Animal):
    def sound(self):
        return "Bark"

print("\nPolymorphism – Animal/Dog:", Dog().sound())
```

```python
abstraction
```

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        return "Bark"

print("\nAbstraction – Animal/Dog:", Dog().sound())
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        return "Car starts"

print("\nAbstraction – Vehicle/Car:", Car().start())
```

```python
from abc import ABC, abstractmethod

class Animal2(ABC):
    @abstractmethod
    def sound(self):
        pass

class Dog2(Animal2):
    def sound(self):
        return "Bark"

print("\nAbstraction – Animal2/Dog2:", Dog2().sound())
```

```python
multiple inheritance
```

```python
class Father:
    def skills(self):
        return "Gardening"

class Mother:
    def traits(self):
        return "Cooking"

class Child(Father, Mother):
    pass

c = Child()
print(f"Child inherits: {c.skills()} and {c.traits()}")
```

```python
multilevel inheritance
```

```python
class Grandfather:
    def property(self):
        return "Land"

class Father(Grandfather):
    def house(self):
        return "House"

class Son(Father):
    pass

s = Son()
print(f"Son inherits: {s.property()} and {s.house()}")
```

```python
encapsulation
```

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def get_balance(self):
        return self.__balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

acc = BankAccount(100)
acc.deposit(50)
print("Balance:", acc.get_balance())
```

```python
class Car:
    def __init__(self, speed):
        self.__max_speed = speed

    def drive(self):
        print(f"Driving at max speed of {self.__max_speed} km/h")

my_car = Car(200)
my_car.drive()
```

```python
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

p = Person("Alice")
print("Person's name:", p.name)
```
